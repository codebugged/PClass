```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,'o:c')
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,'o',ms=10)
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,'o',ms=10,mec='r',mfc='g')
plt.show()
```

```python
x=np.array([3,8,10,11])
y=np.array([2,5,8,11])
plt.plot(x,y)
plt.title("Physics Data")
plt.xlabel("Average Speed")
plt.ylabel("Average Time")

plt.grid()
plt.show()
```

```python
x=np.array([3,8,10,11])
y=np.array([2,5,8,11])
plt.plot(x,y,label="time/speed")
plt.title("Physics Data")
plt.xlabel("Average Speed")
plt.ylabel("Average Time")
plt.axis('equal')
lg = plt.legend()
plt.grid()
plt.show()
```

```python
x=[3,8,10,11]
y=[2,5,8,11]
plt.plot(x,y,label="time/speed")
plt.title("Physics Data")
plt.xlabel("Average Speed")
plt.ylabel("Average Time")
plt.axis('equal')
lg = plt.legend()
plt.grid()
plt.show()
```

```python
figure, axes = plt.subplots()
drawing_colored_circle = plt.Circle((0.6,0.6),0.2,fill=False)
axes.set_aspect(1)
axes.add_artist(drawing_colored_circle)
plt.show()
```

```python
figure, axes = plt.subplots()
axes.set_aspect(1)
plt.xlim(-1,1)
plt.ylim(-1,1)
c1 = plt.Circle((0,0),0.2,color='r',fill=False)
c2 = plt.Circle((0,0),0.4,color='g',fill=False)
axes.add_artist(c1)
axes.add_artist(c2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
# create 1000 equally spaced points between -10 and 10
x = np.linspace(-10, 10, 1000)
plt.xlim(-10,10)
plt.ylim(-10,10)
# calculate the y value for each element of the x vector
y = (4-x*x)**0.5 
ax.plot(x, y)
```
